## Histogram

```python
menu.cook_time.describe()
# menu.cook_time.nunique()
```

```python
cook_time = menu['cook_time']             
cook_time.plot.hist(bins = 100)    
plt.legend()
plt.show()
```

```python
#make an array of integers, in 20-step increments, between 0 and the cook_time.max(), which returns the entry with the highest number.
# Construct histogram plot with 50 bins
cook_time.plot.hist(bins=30)

# Modify X-Axis ticks
plt.xticks(np.arange(0, cook_time.max(), 24))
plt.xticks(rotation = 90) 

plt.legend()
plt.show()
```

```python
# Extracting relevant data
cook_time = menu['cook_time']
prep_time = menu['prep_time']

# Alpha indicates the opacity from 0..1
prep_time.plot.hist(alpha = 0.5 , bins = 50) 
cook_time.plot.hist(alpha = 0.2, bins = 50)

plt.xticks(np.arange(0, cook_time.max(), 40))
plt.xticks(rotation = 45) 
plt.legend()
plt.show()
```

```python
prep_time.plot.hist(alpha = 0.6 , color = 'green', title = 'Cooking time', grid = True, bins = 50)
cook_time.plot.hist(alpha = 0.6, color = 'red', figsize = (7,7), grid = True, bins = 20)

plt.xticks(np.arange(0, cook_time.max(), 40))
plt.xticks(rotation = 45) 

plt.legend()
plt.show()
```

## Area Chart

```python
menu.prep_time.describe()
time = menu.groupby('prep_time').mean()
print(time)
```

```python
time.plot.area()
plt.legend()
plt.show()
```

```python
menu_1=menu[(menu.cook_time!=-1) & (menu.prep_time!=-1)]
#print(menu_1.info())

menu_1.plot.area()

plt.legend()
plt.show()
```

## Pie Chart

```python
flavors = menu["flavor_profile"]

  flavors.unique()
```

```python
flavors = menu[menu.flavor_profile != '-1']

flavors['flavor_profile'].value_counts().plot.pie()

plt.legend()
plt.show()
```

```python
states = (menu[menu.state != '-1'])['state'].value_counts().head(7)
print(states)
# Colors to circle through
colors = ['lightpink','pink','fuchsia','mistyrose','hotpink','deeppink','magenta']
explode = (0, 0, 0.2, 0,0,0,0.7)
states.plot.pie(colors = colors, explode=explode,shadow = True, startangle = 15, title = "States")
plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()
```

```python
df = sns.load_dataset('iris')
sns.distplot(df['petal_length'],kde = False) #(KDE) is a way to estimate the probability density function of a continuous random variable
```

```python
sns.set_context('paper')
 
# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.barplot(x = 'embark_town', y = 'age', data = titanic,
            palette = 'PuRd',ci=None
            )

plt.show()
print(titanic.columns)
```

```python
sns.barplot(x = 'sex', y = 'survived', hue = 'class', data = titanic,
            palette = 'PuRd',
            order = ['male', 'female'],  
                    
                         
            
            ci = 'sd'  
            )
plt.legend()
plt.show()
```

```python
sns.set_context('paper')
sns.barplot(x = 'age', y = 'embark_town', data = titanic,
            palette = 'PuRd', orient = 'h',
            )
plt.show()
```

```python
# create plot
sns.countplot(x = 'class', hue = 'who', data = titanic, palette = 'magma')
plt.title('Survivors')
plt.show()
```
