## Combining & Merging

```python
#Merging
df1 = pd.DataFrame({'abc': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df2 = pd.DataFrame({'abc': ['a', 'b', 'd'],'data2': range(3)})
print(df1)
print(df2)
```

```python
"""This is an example of a many-to-one join; the data in df1 has multiple rows labeled a
and b, whereas df2 has only one row for each value in the key column"""
print(pd.merge(df1, df2,how="right"))
```

```python
pd.merge(df1, df2, on='abc')
```

```python
"""If the column names are different in each object, you can specify them separately"""
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df4 = pd.DataFrame({'lkey': ['a', 'b', 'd'],'data2': range(3)})
print(pd.merge(df3, df4, left_on='lkey', right_on='rkey'))
```

```python
pd.merge(df1, df2, how='outer')
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                      'key2': ['one', 'two', 'one'],
                      'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})
print(left)
print(right)
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```

```python
#Merging on index
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'], 'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
print(left1)
print(right1)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```
