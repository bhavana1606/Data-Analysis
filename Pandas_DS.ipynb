# Series

```python
import pandas as pd
```

```python
obj=pd.Series([5,6,7,9,-2])
print(obj.values)
print(obj.index)
print(obj.value_counts)
```

```python
obj2 = pd.Series([4, 7, -5, 3], index=['d', 'b', 'a', 'c'])
print(obj2)
print(obj2.index)
obj2['d']=6
print(obj2)
print(obj2[['c', 'a', 'd']])
```

```python
print(obj2[obj2>3])
obj2=obj2*3
print(obj2)
print('e' in obj2) #in is membership operator, it check for existence, returns boolean
```

```python
#Create series from Dictionary
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
print(sdata)
obj3=pd.Series(sdata)
print(obj3)
```

```python
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj4 = pd.Series(sdata, index=states)
print(obj4)
```

```python
print(pd.isnull(obj4))
print(pd.notnull(obj4))
#Series method
print(obj4.notnull())
```

```python
print(obj3 + obj4)
```

```python
obj4.name = 'population'
obj4.index.name = 'state'
print(obj4)
```

# DataFrame

```python
data = {'state data': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
'year': [2000, 2001, 2002, 2001, 2002, 2003],
'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
frame=pd.DataFrame(data)
print(frame.head(2))
print(frame.head(-2))  # Not showing last 2
print(frame.tail(2))
print(frame.tail(-2))  # Not showing first 2
```

```python
frame2=pd.DataFrame(data, columns=['pop', 'state data', 'year','abc'],index=['one', 'two', 'three', 'four','five', 'six'])
print(frame2.columns)
```

```python
import numpy as np
frame2['abc'] =16.5
frame2['abc'] = np.arange(6)
frame2['abc'] = [2,3,5,8,9,7]
print(frame2)
val = pd.Series([-1.2, -1.5, -1.7], index=['two', 'four', 'five'])
frame2['abc'] = val
print(frame2)
```

```python
print(frame2['year'])
print(frame2.year)   # after . column name must be valid variable name 'state data' will not work
```

```python
frame2['eastern'] = frame2['state data'] == 'Ohio'
print(frame2)
del frame2['eastern'] 
print(frame2.columns)
```

```python
import pandas as pd
#Nested Dictionary
pop = {'Nevada': {2001: 2.4, 2002: 2.9},'Ohio': {2000: 1.5, 2001: 1.7, 2002: 3.6}}
nFrame=pd.DataFrame(pop)
print(nFrame)
```

```python
print(nFrame.T)
```

```python
frame3=pd.DataFrame(pop, index=[2001, 2002, 2000])
pdata = {'Ohio': frame3['Ohio'][:2],'Nevada': frame3['Nevada'][:2]}
pd.DataFrame(pdata)
frame3.index.name = 'year'; frame3.columns.name = 'state'
print(frame3)
print(frame3.values)
```

# Essential Functionalities

```python
#Reindexing  (Return new onject after reindexing)
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
obj
```

```python
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e'])
obj2
```

```python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 4, 5])
print(obj3)
obj=obj3.reindex(range(6), method='ffill')
print(obj)
```

```python
import numpy as np
frame = pd.DataFrame(np.arange(9).reshape((3, 3)),
                     index=['a', 'c', 'd'], 
                     columns= ['Ohio', 'Texas', 'California'])
print(frame)
frame2 = frame.reindex(['a', 'b'])
print(frame2)
states = ['Texas', 'Utah']
frame2=frame.reindex(columns=states)
print(frame2)
```

```python
#Drop
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
obj
obj2=obj.drop(['a','c'])
print(obj2)
```
